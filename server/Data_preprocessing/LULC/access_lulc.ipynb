```python
import rasterio
import matplotlib.pyplot as plt

# Path to your clipped TIFF file
file_path = r"C:\Users\Ankit\Downloads\lulc_maps_tif\LULC_2015_clipped.tif"

# Open the raster
with rasterio.open(file_path) as src:
    data = src.read(1)  # Read the first band
    
    # Plot the raster
    plt.figure(figsize=(10, 8))
    plt.imshow(data, cmap="tab20", interpolation="nearest")
    plt.title("LULC 2015 (Clipped)")
    plt.colorbar(label="Land Cover Class")
    plt.xlabel("Column Index")
    plt.ylabel("Row Index")
    plt.grid(False)
    plt.show()
    
    # Optional: print metadata
    print("CRS:", src.crs)
    print("Bounds:", src.bounds)
    print("Resolution:", src.res)
```
